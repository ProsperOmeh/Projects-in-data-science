# Extracting the name of plants and their price from website

```python
# importing required libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# target URL to scrap
geturl = "https://www.pepperfry.com/home-decor-vases-flower-indoor-plants.html?type=hover-decor"
```

```python
# headers
headers = {
    'User-Agent': "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.90 Safari/537.36"
    }
```

```python
# send request to download the data
response = requests.request("GET", geturl, headers=headers)
```

```python
# parse the downloaded data
data = BeautifulSoup(response.text, 'html.parser')
print(data.prettify())
```

```python
#find all the sections with specifiedd class name
plant_details = data.find_all('tr', attrs={'class': "seo_text_row seo-text-border-gry seo-pricetag-gry pf-margin-top10"})
```

```python
print('Total Number of plants found: ', len(plant_details))
```

```python
for plants in plant_details:
    print(plants)
```

```python
# extract the plant name and price.

for plants in plant_details:
    plant_name = plants.find('a')
    plant_price = plants.find('td', attrs={'class': 'seo-text-col seo-text-right'})
    print(plant_name.text,plant_price.text)
    
```

# Storing the data


```python
# First we created a list - list_plant and then initialized dictionary 'dataframe' and stored the dictionary in list.

list_plant =[]
for plants in plant_details:
    dataframe ={}
    dataframe['plant_name'] = plants.find('a').text
    dataframe['plant_price'] = plants.find('td', attrs={'class': 'seo-text-col seo-text-right'}).text
    list_plant.append(dataframe)
list_plant
```

```python
# Saving the dictionary to dataframe.
import pandas
df = pandas.DataFrame(list_plant)
df
```

```python
#Finally saving the data to CSV file.
df.to_csv("list_plant.csv",index=False)
```
