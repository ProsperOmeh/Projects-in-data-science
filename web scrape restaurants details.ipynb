# Find top 10 restautants in bangalore

```python
# importing required libraries

import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# target URL to scrap

geturl = "https://www.dineout.co.in/top-10-restaurants-in-bangalore"
```

```python
# headers
headers = {
    'User-Agent': "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.90 Safari/537.36"
    }
```

```python
# send request to download the data
response = requests.request("GET", geturl, headers=headers)
```

```python
# parse the downloaded data
data = BeautifulSoup(response.text, 'html.parser')
print(data.prettify())
```

```python
#find all the sections with specified class name
rest_details = data.find_all('div', attrs={'class': "restnt-main-wrap clearfix"})
```

```python
print('Total count found: ', len(rest_details))
```

```python
for res in rest_details:
    print(res.prettify())
```

```python
# Extracting the below restaurant details:

for res in rest_details:
    
    rest_name = res.find('a')
    rest_address = res.find('div', attrs={'class': 'restnt-loc ellipsis'})
    rest_price = res.find('span', attrs={'class': 'double-line-ellipsis'})
    rest_offer = res.find('li',attrs={'class': 'ellipsis'})
    
    print(rest_name.text,rest_address.text,rest_price.text,rest_offer.text)
    
```

# Storing the data

```python
# First we created a list - list_plant and then initialized dictionary 'dataframe' and stored the dictionary in list.

restaurant_details =[]
for res in rest_details:
    dataframe ={}
    dataframe['rest_name'] = res.find('a').text
    dataframe['rest_address'] = res.find('div', attrs={'class': 'restnt-loc ellipsis'}).text
    dataframe['rest_price'] = res.find('span', attrs={'class': 'double-line-ellipsis'}).text
    dataframe['rest_offer'] = res.find('li',attrs={'class': 'ellipsis'}).text
    restaurant_details.append(dataframe)
restaurant_details
  
```

```python
# Saving the dictionary to dataframe.

import pandas
df = pandas.DataFrame(restaurant_details)
df
```

```python
#Finally saving the data to CSV file.

df.to_csv("restaurant_details.csv",index=False)
```
